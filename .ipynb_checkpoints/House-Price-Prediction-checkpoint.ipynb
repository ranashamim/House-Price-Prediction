```python
import pandas as pd
import matplotlib.pyplot as plt 
import sys
sys.path.append('/home/hooshmand/Documents/Git/myenv/lib/python3.11/site-packages')
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import VarianceThreshold
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv('data/data.csv')
```

```python
data.head(10)
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data = data.drop(['street', 'date', 'statezip', 'country'], axis=1)
y = data['price'].to_numpy()
data = data.drop('price', axis=1)
```

```python
for i in data:
    print(f" {i} : {data[i].unique()} . size : {len(data[i].unique())}")
```

```python
y   
```

```python
features = ['bedrooms', 'sqft_living', 'yr_built', 'city']
markers = [">", "s", "P", "D"]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
```

```python
for i in range(0,len(features)):
    plt.scatter(data[features[i]], y, s=50, c=colors[i], marker=markers[i])
    plt.xticks(rotation=90) 
    plt.xlabel(features[i])
    plt.ylabel('Price')
    plt.show()
    
```

```python
# classify
data['city'] = pd.Categorical(data['city'])
data['city'] = data['city'].cat.codes
```

```python
# normalize
scaleCol = ['bathrooms','sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement']
scalee = data[scaleCol].to_numpy()
for i in range(scalee.shape[1]):
    scaler = StandardScaler()
    scaler.fit(scalee[:,[i]])
    data[scaleCol[i]] = scaler.transform(scalee[:,[i]])
        
```

```python
data
```

```python
data.shape
```

```python
selection = VarianceThreshold(threshold=(.8 * (1 - .8)))
X = selection.fit_transform(data)
X.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X , y, test_size=0.2, random_state=40)
```

```python
model = LinearRegression()
model.fit(X=X_train, y = Y_train)
```

```python
yPred = model.predict(X_test)
```

```python
print(f"Final Score for this Dataset is : %{r2_score(y_true=Y_test, y_pred=yPred) * 100}")
Y_train.shape
```
